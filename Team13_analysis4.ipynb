```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movie=pd.read_csv("2016년여가겨일별데이터.csv", engine='python')
movie=movie.dropna()
```

```python
movie['날짜'] = movie['날짜'].apply(lambda x:pd.to_datetime(str(x), format='%Y%m%d'))
movie
```

```python
movie=movie.reset_index()
movie=pd.DataFrame(movie, columns=['영화명','개봉일','날짜','매출액점유율','관객수','누적관객수'])
```

```python
movie
```

```python
movie.to_csv('2016년여가겨일별데이터_정리.csv')
```

```python
movie=pd.read_csv("2016년여가겨일별데이터_정리.csv", engine='python')
movie=movie.dropna()
```

```python
movie
```

```python
movie['개봉후일수']=pd.to_numeric(movie['개봉후일수'], errors='coerce')
movie=movie.dropna()
movie=movie.where(movie['개봉후일수']>=0).dropna()
movie=movie.where(movie['개봉후일수']<21).dropna()

movie
```

```python
movie_peo=pd.DataFrame(movie['누적관객수'].groupby(movie['영화명']).max())
movie_peo=movie_peo.reset_index()
movie_peo=movie_peo.where(movie_peo['누적관객수']>10000).dropna()
movie_peo
```

```python
mergemovie=pd.merge(movie_peo,movie,on='영화명')
```

```python
mergemovie
```

```python
mergemovie.to_csv('2016년여가겨일별데이터_정리_정리.csv')
```
