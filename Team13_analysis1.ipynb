```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#movie : 일별데이터 데이터셋 2016 여름,겨울,가을 2013 여름,겨울 데이터 포함
#필요한 columns만 수집
movie=pd.read_csv("최종movie.csv", engine='python')
movie=pd.DataFrame(movie,columns=['날짜','순위','영화명','감독','개봉일','관객수','국적','등급','장르','개봉후일수','매출액점유율','누적관객수','스크린수','상영횟수'])
movie.columns=['상영날짜','순위','영화명','감독','개봉일','관객수','국적','등급','장르','개봉후일수','매출액점유율','누적관객수','스크린수','상영횟수']
movie
```

```python
#mvoie_peo : 각 영화의 총관객수와 개봉일을 저장(2004~2017년)
movie_peo=pd.read_csv("영화_누적관객수.csv", engine='python')
movie_peo.columns=['순번','영화명','개봉일','총관객수']
movie_peo=pd.DataFrame(movie_peo,columns=['영화명','총관객수','개봉일'])
movie_peo
```

```python
#데이터 전처리 => 각 데이터마다 총관객수정보 추가
movie=pd.merge(movie,movie_peo,left_on=['영화명','개봉일'],right_on=['영화명','개봉일'])
movie=movie.dropna()
movie
```

```python
#movie_high : 매출액 점유율이 5%를 한번이라도 넘은적이 있는 영화들
#movie_drop : movie_high에 속해있는 영화들의 데이터 중 개봉후 일수가 0~21사이인 데이터
movie_high=movie.where(movie['매출액점유율']>0.05).dropna()
movie_high=pd.DataFrame(movie_high,columns=['영화명'])
movie_high=movie_high.drop_duplicates()
movie['개봉후일수']=pd.to_numeric(movie['개봉후일수'], errors='coerce')
movie=movie.dropna()
movie=movie.where(movie['개봉후일수']>=0).dropna()
movie_drop=movie.where(movie['개봉후일수']<=21).dropna()
movie_drop=pd.merge(movie_high,movie_drop,on='영화명')
movie_drop
```

```python
#장르를 맨앞의 하나로 골라냄
for i in range(len(movie_drop)):
    jan=movie_drop['장르'][i]
    k=0
    for j in range(len(jan)):
        if jan[j]==',':
            k=j
            break
    if k==0:
        k=len(jan)
    movie_drop['장르'][i]=jan[0:k]
```

```python
#장르별로 누적관객/총관객이 70%가 되는 개봉 후 일수 평균
jan_total=movie_drop['영화명'].groupby(movie_drop['장르']).count()
jan_total=jan_total.reset_index()
jan_total=pd.DataFrame(jan_total,columns=['장르'])
jan_=[]
jan_rate=[]
for i in jan_total['장르']:
    movie_drop1=movie_drop.where(movie_drop['장르']==i).dropna()
    movie_drop1['누적관객/총관객']=movie_drop1['누적관객수']/movie_drop1['총관객수']
    movie_drop1=movie_drop1.where(movie_drop1['누적관객/총관객']<0.7).dropna()
    movie_drop1=movie_drop1.where(movie_drop1['누적관객/총관객']>0.6).dropna()
    movie_drop2=pd.DataFrame(movie_drop1['개봉후일수'].groupby(movie_drop1['영화명']).max())
    if len(movie_drop2)>0:
        #print(sum(movie_drop2['개봉후일수'])/len(movie_drop2))
        jan_.append(i)
        jan_rate.append(sum(movie_drop2['개봉후일수'])/len(movie_drop2))
pd.DataFrame({"장르":jan_,"장르별일수":jan_rate})
```

```python
#개봉후 일수에 따른 매출액 점유율 평균
movie_high_merge=pd.merge(movie_high,movie,on='영화명')
date_count=pd.DataFrame(movie_high_merge['매출액점유율'].groupby(movie_high_merge['개봉후일수']).count())
date_mean=pd.DataFrame(movie_high_merge['매출액점유율'].groupby(movie_high_merge['개봉후일수']).mean())
date_count=date_count.where(date_count['매출액점유율']>3).dropna()
result=pd.merge(date_count,date_mean,right_index=True,left_index=True)
#result.to_csv('매출액점유율.csv')
result.columns=['갯수','매출액점유율평균']
result=result.head(22)
result.drop(['갯수'],axis=1,inplace=True)
result.plot(figsize=((20,10)),fontsize=14)
```

```python
#개봉영화에 따른 기존영화의 이전주점유율 다음주 점유율계산
import datetime as dt
movie_start_2016=movie_start.where(movie_start['상영날짜']>'2016-05-31').dropna()
movie_start_2013_sum=movie_start.where(movie_start['상영날짜']>'2013-05-31').dropna()
movie_start_2013_sum=movie_start_2013_sum.where(movie_start['상영날짜']<'2013-09-01').dropna()
movie_start_2013_win=movie_start.where(movie_start['상영날짜']>'2013-11-31').dropna()
movie_start_2013_win=movie_start_2013_win.where(movie_start['상영날짜']<'2014-03-01').dropna()
movie_start=pd.concat([movie_start_2016,movie_start_2013_win,movie_start_2013_sum])
movie_start=movie_start.reset_index()
movie_start=pd.DataFrame(movie_start,columns=['개봉영화','상영날짜','개봉장르','개봉누적관객수'])
result_final=pd.DataFrame(columns=['순위','영화명','누적관객수','총관객수','이전주평균점유율','다음주평균점유율','상영날짜','점유율 변화','장르','개봉후일수'])
for k in range(len(movie_start)):
    inp=movie_start['상영날짜'][k]
    day_list_bf=[]
    day_list_af=[]
    year=int(inp[0:4])
    month=int(inp[5:7])
    day=int(inp[8:])

    for i in range(7):
        date=dt.date(year,month,day)
        afterday=date+dt.timedelta(-i-1)
        new_year=afterday.year
        new_month=afterday.month
        new_day=afterday.day

        result=str(new_year)+'-'
        if new_month<10:
            result=result+'0'+str(new_month)+'-'
        else:
            result=result+str(new_month)+'-'
        if new_day<10:
            result=result+'0'+str(new_day)
        else:
            result=result+str(new_day)
        day_list_bf.append(result)
    
    for i in range(7):
        date=dt.date(year,month,day)
        afterday=date+dt.timedelta(i)
        new_year=afterday.year
        new_month=afterday.month
        new_day=afterday.day

        result=str(new_year)+'-'
        if new_month<10:
            result=result+'0'+str(new_month)+'-'
        else:
            result=result+str(new_month)+'-'
        if new_day<10:
            result=result+'0'+str(new_day)
        else:
            result=result+str(new_day)
        day_list_af.append(result)
    
    total_result=[]
    total_movie=[]
    movie_st=movie_start['상영날짜'][k]
    movie_name=movie_start['개봉영화'][k]
    #movie_gan=movie_start['장르'][k]
    movie_day=movie_drop.where(movie_drop['상영날짜']==movie_st).dropna()
    
    for i in movie_day['영화명']:
        if i!=movie_name:
            total=0
            num=0
            hk=movie_drop.where(movie_drop['영화명']==str(i)).dropna()
            for j in range(7):
                h=hk.where(hk['상영날짜']==day_list_bf[j]).dropna()
                h=h.reset_index()
                h=pd.DataFrame(h,columns=['매출액점유율','개봉후일수'])
                if len(h)>0:
                    after_st=h['개봉후일수'][0]
                    if after_st<7:
                        total+=h['매출액점유율'][0]
                    elif after_st<14:
                        total+=(h['매출액점유율'][0]*1.697)
                    else:
                        total+=(h['매출액점유율'][0]*3.34)
                    num+=1
            if num>0:
                total=total/num
            if total>0:
                total_result.append(total)
                total_movie.append(str(i))
    avg_result=pd.DataFrame({'영화명':total_movie,'이전주평균점유율':total_result})
    
    total_result_af=[]
    total_movie_af=[]
    for i in movie_day['영화명']:
        if i!=movie_name:
            total=0
            num=0
            for j in range(7):
                hk=movie_drop.where(movie_drop['영화명']==str(i)).dropna()
                h=hk.where(hk['상영날짜']==day_list_af[j]).dropna()
                h=h.reset_index()
                h=pd.DataFrame(h,columns=['매출액점유율','개봉후일수'])
                if len(h)>0:
                    after_st=h['개봉후일수'][0]
                    if after_st<7:
                        total+=h['매출액점유율'][0]
                    elif after_st<14:
                        total+=(h['매출액점유율'][0]*1.697)
                    else:
                        total+=(h['매출액점유율'][0]*3.34)
                    num+=1
            if num>0:
                total=total/num
            if total>0:
                total_result_af.append(total)
                total_movie_af.append(str(i))
    avg_result_af=pd.DataFrame({'영화명':total_movie_af,'다음주평균점유율':total_result_af})
    
    
    movie_d=movie_drop.where(movie_drop['상영날짜']==inp).dropna()
    movie_d=pd.merge(avg_result_af,movie_d,on='영화명')
    movie_r=pd.merge(avg_result,movie_d,on='영화명')
    movie_r['점유율 변화']=((movie_r['다음주평균점유율']-movie_r['이전주평균점유율'])/movie_r['이전주평균점유율'])*100
    movie_r=pd.DataFrame(movie_r,columns=['순위','영화명','누적관객수','총관객수','이전주평균점유율','다음주평균점유율','상영날짜','점유율 변화','장르','개봉후일수'])
    #t_movie=pd.DataFrame({'영화명':[movie_name],'이전주평균점유율':[0],'다음주평균점유율':[0],'상영날짜':[movie_st],'장르':[movie_gan]})
    #result_final=pd.concat([result_final,t_movie])                      
    result_final=pd.concat([result_final,movie_r])
result_final
```
