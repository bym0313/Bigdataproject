```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movie=pd.read_csv("2016년여가겨일별데이터_정리.csv", engine='python')
movie=movie.dropna()
```

```python
movie_peo=pd.DataFrame(movie['누적관객수'].groupby(movie['영화명']).max())
movie_peo=movie_peo.reset_index()
movie_peo=movie_peo.where(movie_peo['누적관객수']>10000).dropna()
movie_peo
```

```python
mergemovie=pd.merge(movie_peo,movie,on='영화명')
```

```python
mergemovie
```

```python
movie_name=pd.DataFrame(mergemovie['영화명'])
movie_name=movie_name.drop_duplicates()
movie_name
```

```python
mergemovie['개봉후일수']=pd.to_numeric(mergemovie['개봉후일수'], errors='coerce')
mergemovie=mergemovie.dropna()
mergemovie=mergemovie.where(mergemovie['개봉후일수']>=0).dropna()
mergemovie=mergemovie.where(mergemovie['개봉후일수']<14).dropna()
mergemovie
```

```python
movie_name=pd.DataFrame(mergemovie['영화명'])
movie_name=movie_name.drop_duplicates()
movie_name
```

```python
mergemovie['누적관객수_x']=pd.to_numeric(mergemovie['누적관객수_x'], errors='coerce')
mergemovie=mergemovie.dropna()
movie300=mergemovie.where(mergemovie['누적관객수_x']>=3000000).dropna()
movie50=mergemovie.where(mergemovie['누적관객수_x']<3000000).dropna()
movie50=mergemovie.where(mergemovie['누적관객수_x']>=500000).dropna()
movieextra=mergemovie.where(mergemovie['누적관객수_x']<500000).dropna()
```

```python
movienameextra=pd.DataFrame(movieextra['영화명'])
movienameextra=movienameextra.drop_duplicates()
movienameextra
```

```python
movie300['비율']=movie300['누적관객수_y']/movie300['누적관객수_x']
movie50['비율']=movie50['누적관객수_y']/movie50['누적관객수_x']
movieextra['비율']=movieextra['누적관객수_y']/movieextra['누적관객수_x']
```

```python
movie300=movie300.dropna()
movie300_1=movie300.where(movie300['개봉후일수']<7).dropna()
movie300_2=movie300.where(movie300['개봉후일수']>=7).dropna()

movie50=movie50.dropna()
movie50_1=movie50.where(movie50['개봉후일수']<7).dropna()
movie50_2=movie50.where(movie50['개봉후일수']>=7).dropna()

movieextra=movieextra.dropna()
movieextra_1=movieextra.where(movieextra['개봉후일수']<7).dropna()
movieextra_2=movieextra.where(movieextra['개봉후일수']>=7).dropna()
```
