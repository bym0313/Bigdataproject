```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rc('font', family="HYGothic-Extra")
```

```python
movie=pd.read_csv("movie3.csv", engine='python')
movie
```

```python
movie_drop=movie.fillna(0)
movie_drop
```

```python
movie_drop['계절']=''
for i in range(len(movie_drop)):
    달=movie_drop['달'][i]
    if 달>2 and 달<6:
        movie_drop['계절'][i]='봄'
    elif 달>5 and 달<9:
        movie_drop['계절'][i]='여름'
    elif 달>8 and 달<12:
        movie_drop['계절'][i]='가을'
    else:
        movie_drop['계절'][i]='겨울'
movie_drop
```

```python
for i in range(len(movie_drop)):
    jan=movie_drop['장르'][i]
    k=0
    for j in range(len(jan)):
        if jan[j]==',':
            k=j
            break
    if k==0:
        k=len(jan)
    movie_drop['장르'][i]=jan[0:k]
movie_drop
```

```python
movie_nondu=pd.DataFrame(movie_drop['누적관객수'].groupby(movie_drop['영화명']).max()).sort_values(by=['누적관객수'],ascending=[False])
movie_nondu=movie_nondu.reset_index()
movie_drop=pd.merge(movie_drop,movie_nondu,on=['영화명','누적관객수'])
movie_drop=movie_drop.where(movie_drop['년도']>=2004).dropna()
movie_drop
```

```python
#장르-누적관객수(합)
pd.DataFrame(movie_drop['누적관객수'].groupby(movie_drop['장르']).sum()).sort_values(by=['누적관객수'],ascending=[False])
```
